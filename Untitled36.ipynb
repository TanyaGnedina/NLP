тут же решим проблему с разными обозначениями в разных методах, создадим функцию, приводящую все метки к виду, который используется в нашей разметке.

```python
DICT_of_pos= {'NOUN':{'NN', 'NOUN', 'PROPN', 'S', 'NNS', 'NNP'},
              'VERB':{'V', 'VERB', 'INFN', 'PRTF', 'AUX', 'PRTS','GRND','VB', 'VBP', 'VBD', 'VBG', 'VBZ', 'PRED', 'MD'},
              'ADV':{'ADVB', 'ADV', 'ADVPRO', 'RB'},
              'ADJ':{'ADJF', 'ADJ', 'A', 'ADJS', 'ADJF', 'COMP', 'JJ'},
              'PRON':{'APRO','SPRO', 'NPRO', 'DET','PRON', 'PRP'},
              'PREP':{'ADP', 'PR', 'PREP', 'IN','RP'},
              'PART':{'PART', 'PRCL', 'TO'},
              'CONJ':{'CCONJ', 'CONJ', 'SCONJ', 'CC', 'INTJ'},
              'DET':{'DT', 'DET', 'PRP$'},
              'PUNCT':{'.',',','-'}}
```

```python
def common_tags(list1):
    for i in list1:
        for j in DICT_of_pos:
            if i[1] in DICT_of_pos[j]:
                i[1]=j
    return(list1)
```

## Accuracy 

```python
def accuracy(one_of, gold):
    count_equal = 0
    for z in range(len(one_of)):
        if one_of[z][1]==gold[z][1]:
            count_equal += 1
    return(count_equal/len(one_of))
```
